```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dat_swin = np.load("./swin_roc.npy")
swin_auc = 0.71704262
```

```python
dat_alex = np.load("./alex_roc.npy")
alex_auc = 0.5
```

```python
plt.title("ROC-кривые моделей")
# SWIN
plt.plot([0,1],[0,1], label="Случайный выбор", linestyle="dashed")
plt.plot(dat_swin[0],dat_swin[1], label=f"SWIN (AUC: {round(swin_auc,3)})")
# AlexNet
plt.plot(dat_alex[0],dat_alex[1], label=f"AlexNet (AUC: {round(alex_auc,3)})")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.legend()
```
